```python
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob

# Path to all CSV files
csv_files = glob("gr_metrics_game_*.csv")

if not csv_files:
    raise FileNotFoundError("No Dynamic_GR CSV files found.")

# Results placeholder (replace with actual win/loss tracking)
results_data = {
    'Game_ID': range(1, len(csv_files) + 1),
    'Result': ['Win'] * 15 + ['Lose'] * 5  # Example: First 40 games are wins, next 60 are losses
}
results_df = pd.DataFrame(results_data)

# Consolidated summary list
summary_list = []

for file in csv_files:
    try:
        # Read the CSV file
        df = pd.read_csv(file)
        game_id = int(file.split("_")[-1].split(".")[0])  # Extract Game ID

        # Skip files with no rows
        if df.empty:
            print(f"Skipping file with no data: {file}")
            continue

        # Ensure required columns are present
        required_columns = ['Step', 'GR', 'Complexity', 'Entropy']
        for col in required_columns:
            if col not in df.columns:
                df[col] = None  # Add missing columns

        # Win/Loss Tracking
        result = results_df.loc[results_df['Game_ID'] == game_id, 'Result'].values[0]

        # Calculate Win/Loss Stats
        win_count = (results_df['Result'] == 'Win').sum()
        lose_count = (results_df['Result'] == 'Lose').sum()
        total_games = len(results_df)
        win_ratio = win_count / total_games

        print(f"Game {game_id} Win/Loss Ratio:")
        print(f"Wins: {win_count}, Losses: {lose_count}, Total: {total_games}")
        print(f"Win Rate: {win_ratio*100:.2f}%")

        # Plot Complexity and Entropy
        fig, ax1 = plt.subplots(figsize=(10, 6))
        color = 'tab:blue'
        ax1.set_xlabel('Step')
        ax1.set_ylabel('Complexity', color=color)
        ax1.plot(df['Step'], df['Complexity'], color=color, label='Complexity')
        ax1.tick_params(axis='y', labelcolor=color)

        ax2 = ax1.twinx()
        color = 'tab:red'
        ax2.set_ylabel('Entropy', color=color)
        ax2.plot(df['Step'], df['Entropy'], color=color, linestyle='--', label='Entropy')
        ax2.tick_params(axis='y', labelcolor=color)

        fig.tight_layout()
        plt.title(f"Complexity and Entropy Over Time (Game {game_id})")
        plt.show()

        # Entropy Analysis
        entropy_start = df['Entropy'].iloc[0]
        entropy_end = df['Entropy'].iloc[-1]
        print(f"Entropy at start: {entropy_start:.2f}")
        print(f"Entropy at end: {entropy_end:.2f}")
        if entropy_end < entropy_start:
            print("Entropy decreased over time, indicating increased certainty.")
        else:
            print("Entropy did not decrease as expected, needs further analysis.")

        # Plot GR Over Time
        plt.figure(figsize=(10, 6))
        plt.plot(df['Step'], df['GR'], label='GR', color='green')
        plt.title(f"GR Value Over Time (Game {game_id})")
        plt.xlabel("Step")
        plt.ylabel("GR")
        plt.grid(True)
        plt.legend()
        plt.show()

        # Check for Significant GR Changes
        gr_diff = df['GR'].diff().fillna(0)
        large_changes = gr_diff.abs() > (df['GR'].mean() * 0.5)
        significant_changes = df[large_changes]

        if not significant_changes.empty:
            print("Significant GR changes at steps:")
            print(significant_changes[['Step', 'GR']])
        else:
            print("No large sudden GR changes detected.")

        # Summary Statistics for the Game
        summary = {
            'Game_ID': game_id,
            'Initial_Complexity': df['Complexity'].iloc[0],
            'Final_Complexity': df['Complexity'].iloc[-1],
            'Initial_Entropy': df['Entropy'].iloc[0],
            'Final_Entropy': df['Entropy'].iloc[-1],
            'Mean_GR': df['GR'].mean(),
            'Max_GR': df['GR'].max(),
            'Min_GR': df['GR'].min(),
            'Result': result
        }
        summary_list.append(summary)

    except Exception as e:
        print(f"Error processing file {file}: {e}")

# Create Summary DataFrame
summary_df = pd.DataFrame(summary_list)

print("\nSummary Statistics Across All Games:")
print(summary_df)

# Save Summary to CSV
summary_df.to_csv("summary_statistics_all_games.csv", index=False)

# Plot Aggregated Metrics Across All Games
plt.figure(figsize=(12, 6))
plt.plot(summary_df['Game_ID'], summary_df['Mean_GR'], label='Mean GR', color='green')
plt.scatter(summary_df['Game_ID'], summary_df['Initial_Entropy'], label='Initial Entropy', color='blue', alpha=0.6)
plt.scatter(summary_df['Game_ID'], summary_df['Final_Entropy'], label='Final Entropy', color='red', alpha=0.6)
plt.title("Aggregated Metrics Across Games")
plt.xlabel("Game ID")
plt.ylabel("Metric Value")
plt.legend()
plt.grid(True)
plt.show()

# Print Games with Highest and Lowest GR
print("Games with Highest GR:")
print(summary_df.nlargest(5, 'Mean_GR')[['Game_ID', 'Mean_GR']])
print("\nGames with Lowest GR:")
print(summary_df.nsmallest(5, 'Mean_GR')[['Game_ID', 'Mean_GR']])
```
